# Candle Plot Demo

```python
import pandas as pd
import utils
import plotly.graph_objects as go
```

```python
pair = "EUR_USD"
granularity = "H1"
```

```python
df = pd.read_pickle(utils.get_his_data_filename(pair, 4000, granularity))
df.describe()
```

```python
df.head()
```

```python
# need to clean the data so that i can turn the strings into numbers
non_cols = ['time', 'volume']
mod_cols = [col for col in df.columns if col not in non_cols]
mod_cols
```

```python
df[mod_cols] = df[mod_cols].apply(pd.to_numeric)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df_plot = df.iloc[-100:].copy()
```

```python
# create plot
fig = go.Figure()
fig.add_trace(
    go.Candlestick(
        x=df_plot['time'], 
        open=df_plot['mid_o'], 
        high=df_plot['mid_h'], 
        low=df_plot['mid_l'], 
        close=df_plot['mid_c'],
        line=dict(width=1.5), opacity=1,
        increasing_fillcolor='#24A06B',
        decreasing_fillcolor="#CC2E3C",
        increasing_line_color='#2EC886',
        decreasing_line_color='#FF3A4C'
    )
)
# add formatting
fig.update_layout(width=1000, height=400, title=f"{pair} {granularity} candles", yaxis_title="Price", xaxis_title="Time", 
                  margin=dict(l=10, r=10, b=10),
                  font=dict(size=10, color="#ffffff"),
                  paper_bgcolor="#1e1e1e", plot_bgcolor="#1e1e1e")

fig.update_xaxes(
    gridcolor="#1f292f",
    showgrid=True, fixedrange=True, rangeslider=dict(visible=False)
)

fig.update_yaxes(gridcolor="#1f292f", showgrid=True)

fig.show()
```
